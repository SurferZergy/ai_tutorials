```python
import ray
from ray import tune
from ray.rllib.algorithms.ppo import PPO

# Initialize Ray
ray.init()

# Configure the environment and trainer
config = {
    "env": "CartPole-v1",  # Environment (CartPole in this case)
    "framework": "torch",      # Specify TensorFlow framework (or use "torch" for PyTorch)
    "num_workers": 1,       # Number of parallel workers for training
    "train_batch_size": 4000,  # Total batch size for training per iteration
    "sgd_minibatch_size": 128,  # Minibatch size for each SGD iteration
    "num_sgd_iter": 10,        # Number of SGD iterations per minibatch
    "lr": 0.0003,            # Learning rate for the optimizer
    "gamma": 0.99,           # Discount factor
}

# Create the PPO algorithm object
trainer = PPO(config=config)

# Train for n iterations
n_iter = 10
for i in range(n_iter):
    result = trainer.train()
    print(f"Iteration: {i + 1}, Result: {result}")

# Save the trained model
checkpoint = trainer.save()
print(f"Checkpoint saved at {checkpoint}")

# Shutdown Ray after training
ray.shutdown()
```
